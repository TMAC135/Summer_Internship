```python
def store_news_contends(urls):
    '''
    this function is used to parse the given url and store news(unicode strings) to a list
    Input: urls, list of all urls we want to parse
    Output: a list with strings, each represent the contends of each new
    '''
    # special cases
    if not urls:
        return None
    
    #过滤掉空格和换行符号
    s_all = []
    for url in urls:
        request = requests.get(url) 
        try:
            html = request.text
        except Exception,e:
            print e.error
        soup = BeautifulSoup(html)
        tmp = ''
        for s in soup.find_all('p',class_='para'):
            for i in s.strings:
                tmp += i
        s_all.append(tmp)
    return s_all


print store_news_contends(spider.urls)
```
